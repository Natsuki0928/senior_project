```python
import numpy as np
from gurobipy import *

def min_max(x, axis=None):
    min = x.min(axis=axis, keepdims=True)
    max = x.max(axis=axis, keepdims=True)
    result = (x-min)/(max-min)
    return result
X= np.array([[1,2,3,4,5],[1,0,0,0,1]]).T
X = min_max(X, axis=0)
y=[15,20,34,42,58]
depth =2
T0 = 2**(depth) -1
T = 2**(depth+1) -1 #tree index t
Branch   = [t for t in range(T0)]
Leaf = [t for t in range(T0,T)]
p = 2 #dim(feature) index j
n = len(y) #data index i
AL, AR = {}, {}
for t in Leaf:
    AL[t]=set([])
    AR[t]=set([])
    i = t
    while i!=0:
        parent = (i-1)//2
        if parent*2==i-1:
            AL[t].add(parent)
        else:
            AR[t].add(parent)
        i = parent
print(T)
print(Branch, Leaf)
print(AL)
print(AR)
m =Model()
#変数
d, a, b, z, l, f, L, beta ={}, {},{},{}, {}, {}, {}, {}
for t in Branch:
    d[t] = m.addVar(name=f"d{t}“, vtype=“B") #ノードtで分枝をするとき１
    b[t] = m.addVar(name=f"b{t}“, vtype=“C") #ノードtでの閾値
for t in Leaf:
    l[t] = m.addVar(name=f"l{t}“, vtype=“B") #葉が空でないとき1
for j in range(p):
    for t in Branch:
        a[j,t] = m.addVar(name=f"a{j}{t}“, vtype=“B") #ノードtが特徴jで分けるとき1
for i in range(n):
    for t in Leaf:
        z[i,t] = m.addVar(name=f"z{i}{t}“, vtype=“B") #データiが葉tに入るとき１
for t in Leaf:
    beta[t] = m.addVar(name=f"beta{t}“, vtype=“C") #葉tの予測値
for i in range(n):
    f[i] = m.addVar(name=f"f{i}“, vtype=“C") #データiの予測値
for i in range(n):
    L[i]  = m.addVar(name = f"L{i}“, vtype=“C" )  #ロス
#制約 (dとa,bの関係）)
for t in Branch:
    m.addConstr( quicksum(a[j,t] for j in range(p)) == d[t])
    m.addConstr(b[t] <= d[t] )
#親子関係
for t in range(1,T0):
    m.addConstr( d[t]<=d[(t-1)//2] )
#データiがノードtに割り当てられたらl[t]が1；葉の最小データ数
for t in Leaf:
    for i in range(n):
        m.addConstr( z[i,t] <= l[t] )
    m.addConstr( quicksum( z[i,t] for i in range(n)) >=1*l[t]) #Lmin
#何れかの葉に入る．
for i in range(n):
    m.addConstr( quicksum(z[i,t] for t in Leaf)  ==1 )
#分岐条件
for t in Leaf:
    for i in range(n):
        for ansestor in AL[t]:
            m.addConstr( quicksum(a[j,ansestor]*X[i][j] for j in range(p)) <= b[ansestor]+ 10*(1-z[i,t])-0.001 )
        for ansestor in AR[t]:
            m.addConstr( quicksum(a[j,ansestor]*X[i][j] for j in range(p)) >= b[ansestor]- 10*(1-z[i,t]) )
#予測値
for t in Leaf:
    for i in range(n):
        m.addConstr( f[i] - beta[t] >= 100*(z[i,t]-1) )
        m.addConstr( f[i] - beta[t] <= 100*(1-z[i,t]) )
#絶対値
for i in range(n):
    m.addConstr( L[i] >= f[i] - y[i] )
    m.addConstr( L[i] >= -f[i] + y[i] )
m.setObjective( quicksum( L[i] for i in range(n) ), GRB.MINIMIZE )
#print(m)
m.optimize()
print(m.Status, m.ObjVal)
for v in m.getVars():
    if v.X >= 0.001:
        print(v.VarName, v.X) 
```
