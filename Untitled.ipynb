```python
import xml.etree.ElementTree as ET
import pandas as pd

def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
        all_records.append(record)
    df = pd.DataFrame(all_records)
    return df

# load XML to dataframe (gotta be small)
xml_data = open('data/P20-12_04.xml').read()
df = xml2df(xml_data)
df.tail()
```

```python
depth =2
T0 = 2**(depth) -1
T = 2**(depth+1) -1 #tree index t
Branch   = [t for t in range(T0)]
Leaf = [t for t in range(T0,T)]
p = 2 #dim(feature) index j
K = 2 #classes index k
X = [[0.1,0.1], [0.2,0.7], [0.3,0.4], [0.6,0.6], [0.7, 0.8], [0.6,0.1], [0.7,0.3] ]
y = [0, 0, 0, 0, 0, 1, 1]
n = len(y) #data index i
AL, AR = {}, {}
for t in Leaf:
    AL[t]=set([])
    AR[t]=set([])
    i = t
    while i!=0:
        parent = (i-1)//2
        if parent*2==i-1:
            AL[t].add(parent)
        else:
            AR[t].add(parent)
        i = parent
print(T0,T)
#print(TB, Leaf)
print(AL)
print(AR)
m =Model()
#パラメータYの準備
Y ={}
for i in range(n):
    for k in range(K):
        if y[i]==k:
            Y[i,k] = 1
        else:
            Y[i,k] = -1
#変数
d, a, b, z, l, N, NN, c, L ={}, {},{},{}, {}, {}, {}, {}, {}
for t in Branch:
    d[t] = m.addVar(name=f"d{t}", vtype="B")
    b[t] = m.addVar(name=f"b{t}", vtype="C")
for t in Leaf:
    l[t] = m.addVar(name=f"l{t}", vtype="B")
for j in range(p):
    for t in Branch:
        a[j,t] = m.addVar(name=f"a{j}{t}", vtype="B")
for i in range(n):
    for t in Leaf:
        z[i,t] = m.addVar(name=f"z{i}{t}", vtype="B")
for t in Leaf:
    for k in range(K):
        N[k,t] = m.addVar(name=f"N{k}{t}", vtype="C")
        c[k,t] = m.addVar(name=f"c{k}{t}", vtype="B")
#同じ記号なのでNNにする．
for t in Leaf:
    NN[t] = m.addVar(name = f"NN{t}", vtype="C" )
    L[t]  = m.addVar(name = f"L{t}", vtype="C" )
#制約 (dとa,bの関係）)
for t in Branch:
    m.addConstr( quicksum(a[j,t] for j in range(p)) == d[t])
    m.addConstr(b[t] <= d[t] )
#親子関係
for t in range(1,T0):
    m.addConstr( d[t]<=d[(t-1)//2] )
#データiがノードtに割り当てられたらl[t]が1；葉の最小データ数
for t in Leaf:
    for i in range(n):
        m.addConstr( z[i,t] <= l[t] )
    m.addConstr( quicksum( z[i,t] for i in range(n)) >=1*l[t]) #Lmin
#何れかの葉に入る．
for i in range(n):
    m.addConstr( quicksum(z[i,t] for t in Leaf)  ==1 )
#分岐条件
for t in Leaf:
    for i in range(n):
        for ansestor in AL[t]:
            m.addConstr( quicksum(a[j,ansestor]*X[i][j] for j in range(p)) <= b[ansestor]+ 10*(1-z[i,t])-0.001 )
        for ansestor in AR[t]:
            m.addConstr( quicksum(a[j,ansestor]*X[i][j] for j in range(p)) >= b[ansestor]- 10*(1-z[i,t]) )
#葉に含まれるクラスkの数
for t in Leaf:
    for k in range(K):
        m.addConstr( 2*N[k,t] == quicksum( (1+Y[i,k])*z[i,t] for i in range(n)))
#葉に含まれるデータ数
for t in Leaf:
    m.addConstr(NN[t] ==quicksum( z[i,t] for i in range(n)))
#葉に割り当てられた最大のクラス
for t in Leaf:
    m.addConstr( quicksum(c[k,t] for k in range(K))==l[t])
for t in Leaf:
    for k in range(K):
        m.addConstr(L[t] >= NN[t] -N[k,t]- 10*(1-c[k,t]))
        m.addConstr(L[t] <= NN[t] -N[k,t]+ 10*c[k,t])
m.setObjective( quicksum( L[t] for t in Leaf ), GRB.MINIMIZE )
#print(m)
m.optimize()
print(m.Status, m.ObjV)
```
